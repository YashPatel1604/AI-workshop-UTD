```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sales = pd.read_csv('sales_data.csv', parse_dates = ['Date'])
```

```python
sales.shape
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sales['Unit_Cost'].describe()
```

```python
sales['Unit_Cost'].plot(kind='box', vert = False, figsize=(14,6))
```

```python
sales['Unit_Cost'].plot(kind='density', figsize=(14,6))
```

```python
sales['Unit_Cost'].plot(kind='line', figsize=(14,6))
```

```python
ax = sales['Unit_Cost'].plot(kind='density', figsize=(14,6))
ax.axvline(sales['Unit_Cost'].mean(), color='red')
ax.axvline(sales['Unit_Cost'].median(),color = 'green')
```

```python
ax = sales['Unit_Cost'].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Number of Sales')
ax.set_xlabel('Dollars')
```

```python
sales['Age_Group'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
corr = sales.corr()
corr
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr[:8][:8],cmap = 'RdBu', fignum = fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation = 'vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
sales.plot(kind='scatter',x = 'Revenue', y ='Profit')
```

```python
ax = sales[['Profit', 'Age_Group']].boxplot(by="Age_Group", figsize=(10,6))
ax.set_ylabel('Profit')
```

```python
ax = sales[['Profit', 'Age_Group','Country']].boxplot(by=["Age_Group",'Country'],figsize=(75,10))
ax.set_ylabel('Profit')
```

```python
boxplot_cols = ['Year', 'Customer_Age', 'Order_Quantity','Unit_Cost','Unit_Price','Profit']

sales[boxplot_cols].plot(kind='box',subplots = True, layout=(2,3),figsize=(14,8))

sales[boxplot_cols].plot(kind='box',subplots = False, layout=(2,3),figsize=(14,8))
```

```python
ax = sales[['Profit', 'Age_Group','Country']].groupby(["Age_Group",'Country']).plot(kind='box',figsize=(10,10))
#ax.set_ylabel('Profit')
```

```python
plt.close()
ax = sales[['Profit', 'Age_Group','Country']].groupby(["Age_Group"]).boxplot(by=['Country'],figsize=(50,35))
```
